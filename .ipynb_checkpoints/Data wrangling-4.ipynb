## Town

```python
import pandas as pd

gdf = pd.read_csv('transporte.csv')
gdf
```

```python
tdf = gdf.drop(columns=['id', 'name', 'line', 'geometry', 'longitude', 'latitude', 'type'])
tdf.rename(columns={'town':'name'}, inplace=True)
tdf.drop_duplicates(subset=['name'], inplace=True)
tdf.reset_index(drop=True, inplace=True)
```

```python
# tdf.reindex(columns=['id']+tdf.columns.tolist()+['province', 'latitude', 'longitude', 'address', 'reward'])
```

```python
tdf.to_csv('town.csv', header=True, index=False)
```

### Cambiar tipo town a _uuid_ (id)

```python
tdf = pd.read_csv('TOWN_rows.csv')  # csv con los 'id' correspondientes a cada 'town'
tdf = tdf.drop(columns=['created_at', 'province', 'longitude', 'latitude', 'address', 'reward'])
tdf.rename(columns={'name':'town'}, inplace=True)
tdf
```

```python
import numpy as np

gdf['id'] = np.nan                               # crear columna id
idMap = tdf.set_index('town')['id'].to_dict()    # crear diccionario con la correspondencia entre 'town' e 'id'
gdf['id'] = gdf['town'].map(idMap)               # actualizar 'id'
gdf.drop(columns=['town'], inplace=True)         # quitar 'town'
gdf.rename(columns={'id':'town'}, inplace=True)  # renombrar 'id' a 'town'
gdf
```

```python
gdf.to_csv('station.csv', header=True, index=False)
```
