```python
import json
import numpy as np
import pandas as pd
import re
```

## Reward (premio)

```python
ddf = pd.read_csv('premio_municipio.csv')             # cargar recompensas
ddf.drop(columns=['id', 'created_at'], inplace=True)  # quitar id y created_at
ddf
```

```python
ddf.to_csv('rewardTown.csv', header=True, index=False)  # guardar csv
```

## Town (municipio)

```python
tdf = pd.read_csv('municipio.csv')                      # cargar municipios
tdf.drop(columns=['uuid', 'created_at'], inplace=True)  # quitar uuid y created_at

rid = pd.read_csv('REWARD_rows.csv')                                   # cargar REWARD con los id asociados
rid = rid.drop(columns=['created_at', 'type', 'experience', 'image'])  # quitar columnas no necesarias

# quitar el tipo de recompensa delante del nombre en REWARD para que coincida con los nombres de TOWN
for i in range(len(rid)):
    rid.loc[i, 'name'] = re.sub('emblem = ', '', rid.loc[i, 'name'])
    rid.loc[i, 'name'] = re.sub('medal = ', '', rid.loc[i, 'name'])
    rid.loc[i, 'name'] = re.sub('_L\'Eliana', '', rid.loc[i, 'name'])

dctReward = rid.set_index('name')['id'].to_dict()  # crear diccionario con la correspondencia entre 'name' y 'id' (REWARD)
tdf['reward'] = tdf['name'].map(dctReward)         # actualizar 'reward' ('id' de REWARD) en TOWN

tdf
```

```python
tdf.to_csv('town.csv', header=True, index=False)  # guardar csv
```

## Stop (parada)

```python
sdf = pd.read_csv('Stop - Stop.csv')    # cargar paradas
sdf.drop(columns=['id'], inplace=True)  # quitar id

tid = pd.read_csv('TOWN_rows.csv')                                                                # cargar TOWN con los id asociados
tid = tid.drop(columns=['created_at', 'province', 'latitude', 'longitude', 'address', 'reward'])  # quitamos columnas no necesarias

sdf['id'] = np.nan                               # crear columna 'id'
dctTown = tid.set_index('name')['id'].to_dict()  # crear diccionario con la correspondencia entre 'name' y 'id' (TOWN)
sdf['id'] = sdf['town'].map(dctTown)             # actualizar 'id' en STOP

sdf.drop(columns=['town'], inplace=True)         # quitar 'town'
sdf.rename(columns={'id':'town'}, inplace=True)  # renombrar 'id' a 'town'

# hacemos lo mismo para reward
sdf['reward'] = sdf['name'].map(dctReward)        # actualizar 'reward' en STOP

sdf
```

```python
sdf.to_csv('stop.csv', header=True, index=False)  # guardar csv
```

## Station (estación)

```python
gdf = pd.read_csv('transporte.csv')
gdf.drop(columns=['id'], inplace=True)

nan = gdf.index[gdf.isnull().any(axis=1)]  # miramos si hay faltantes
```

```python
# localizamos faltantes
for i in nan:
    print(gdf.loc[i])
```

```python
# sustituimos los faltantes por los valores correspondientes
gdf.loc[1277, 'line'] = 'C5,C6'
gdf.loc[1286, 'line'] = 'C5,C6'
gdf.loc[1287, 'line'] = 'C1'
gdf.loc[1291, 'line'] = 'C2'
gdf.loc[1294, 'line'] = 'C5'
gdf.loc[1327, 'line'] = 'C2'
```

```python
# vemos que se hayan cambiado correctamente
for i in nan:
    print(gdf.loc[i])
```

```python
# convertir tipo línea a json string
jsonl = {}
for l in gdf['line']:
    jsonl[l] = json.dumps(str(l).split(','))
    
# actualizar valor en el dataframe
gdf['line'] = gdf['line'].replace(jsonl)

gdf['id'] = np.nan                    # crear columna id
gdf['id'] = gdf['town'].map(dctTown)  # actualizar 'id' en STATION (usamos dctTown instanciado arriba)

gdf.drop(columns=['town'], inplace=True)         # quitar 'town'
gdf.rename(columns={'id':'town'}, inplace=True)  # renombrar 'id' a 'town'

gdf
```

```python
gdf.to_csv('station.csv', header=True, index=False)  # guardar csv
```
